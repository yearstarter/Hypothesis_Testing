```python
# Hypothesis:
# H0: there is no difference in click-through rate for the versions
# H1: there is at least one different version
```

```python
# calculating CTR for each version
v1_clicked = int(interact.query('Name == "INTERACT"')["No. clicks"])
v1_notclicked = 10283 - v1_clicked

v2_clicked = int(connect.query('Name == "CONNECT"')["No. clicks"])
v2_notclicked = 2742 - v2_clicked

v3_clicked = int(learn.query('Name == "LEARN"')["No. clicks"])
v3_notclicked = 2747 - v3_clicked

v4_clicked = int(help.query('Name == "HELP"')["No. clicks"])
v4_notclicked =  3180 - v4_clicked

v5_clicked = int(services.query('Name == "SERVICES"')["No. clicks"])
v5_notclicked =  2064 - v5_clicked
```

```python
alpha = 0.1

Click = [v1_clicked, v2_clicked, v3_clicked, v4_clicked, v5_clicked]
No_click = [v1_notclicked, v2_notclicked, v3_notclicked, v4_notclicked, v5_notclicked]

contingency_table = pd.DataFrame(data = [Click, No_click],
                                 columns = ["INTERACT", "CONNECT", "LEARN", "HELP", "SERVICES"],
                                 index = ["Click", "No-click"])
```

```python
chisq, pvalue, df, expected = stats.chi2_contingency(contingency_table)
pvalue
```

```python
pvalue < alpha
```

p-value is less than alpha, therefore we can reject null-hypothesis
